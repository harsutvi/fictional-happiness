```python
#importerer pakker
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Importerer datasett
df = pd.read_csv("https://uit-sok-1006-v23.github.io/data/suits_watermelons.csv", delimiter=";")
```

```python
#Gj√∏r om eksponential fra datasettet
df["p"] = np.exp(df["log p"])
df["h"] = np.exp(df["log h"])
df["q"] = np.exp(df["log q"])

#kontrollsjekk
df
```

```python
#plotter verdiene fra p
plt.plot(df['Year'], df['p'], 'o-')
plt.xlabel('Year')
plt.ylabel('Price')
plt.title('Watermelon market price 1930-1951')
```

```python
#plotter og sammenligner q og h
plt.plot(df['Year'], df['h'], 'o-', label = 'Harvest')
plt.plot(df['Year'], df['q'], 'o--', label = 'Crop')
plt.xlabel('Year')
plt.ylabel('Price')
plt.legend()
```

```python
Q = np.arange(20, 101, 5)
Log_Q = np.log(Q)
Supply_logp = -2+(1.7*Log_Q)
Demand_logp = 8.5-(0.82*Log_Q)
Supply_p = np.exp(Supply_logp)
Demand_p = np.exp(Demand_logp)
New_Supply_logp = Supply_logp+0.4
New_Supply_p = np.exp(New_Supply_logp)



table = pd.DataFrame({
    'Quantity':Q, 
    'Log Quantity':Log_Q, 
    'Supply(log P)':Supply_logp, 
    'Demand(log P)':Demand_logp, 
    'Supply(P)':Supply_p, 
    'Demand(P)':Demand_p, 
    'New_Supply(log_P)':New_Supply_logp,
    'New_Supply(P)':New_Supply_p
})

table
```

```python
plt.plot(Q, Supply_p, 'o-', label='Supply(P)')
plt.plot(Q, Demand_p, 'o-', label='Demand(P)')
plt.legend()
plt.title('Supply and Demand curve for watermelons')
plt.xlabel('Quantity in millions')
plt.ylabel('Price per thousand')
```
